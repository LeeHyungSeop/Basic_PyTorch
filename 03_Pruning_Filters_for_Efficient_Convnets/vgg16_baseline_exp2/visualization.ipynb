```python
import matplotlib.pyplot as plt
import pickle
```

```python
# Load data
with open('./checkpoint/train_acc_list.pkl', 'rb') as f:
    train_acc_list = pickle.load(f)
with open('./checkpoint/train_loss_list.pkl', 'rb') as f:
    train_loss_list = pickle.load(f)
with open('./checkpoint/val_acc_list.pkl', 'rb') as f:
    val_acc_list = pickle.load(f)
with open('./checkpoint/val_loss_list.pkl', 'rb') as f:
    val_loss_list = pickle.load(f)
```

```python
# Plot
plt.figure(figsize=(8, 6))

plt.subplot(2, 1, 1)
plt.plot(train_acc_list, '-', label='train_acc')
plt.plot(val_acc_list, '-', label='val_acc')
plt.title('Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(train_loss_list, '-', label='train_loss')
plt.plot(val_loss_list, '-', label='val_loss')
plt.title('Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.savefig('./visualization.png')
```
