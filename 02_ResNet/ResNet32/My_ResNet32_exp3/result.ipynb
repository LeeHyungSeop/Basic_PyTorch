```python
import pickle
```

```python
# log/top1_acc.pkl
# log/top5_acc.pkl
# log/train_acc_list.pkl
# log/train_loss_list.pkl
# log/val_acc_list.pkl
# log/val_loss_list.pkl

with open('log/top1_acc.pkl', 'rb') as f:
    top1_acc = pickle.load(f)
with open('log/top5_acc.pkl', 'rb') as f:
    top5_acc = pickle.load(f)
with open('log/train_acc_list.pkl', 'rb') as f:
    train_acc_list = pickle.load(f)
with open('log/train_loss_list.pkl', 'rb') as f:
    train_loss_list = pickle.load(f)
with open('log/val_acc_list.pkl', 'rb') as f:
    val_acc_list = pickle.load(f)
with open('log/val_loss_list.pkl', 'rb') as f:
    val_loss_list = pickle.load(f)
    
```

```python
# visualization
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(train_loss_list, label='train')
plt.plot(val_loss_list, label='val')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend()
plt.title('Loss')

plt.subplot(1, 2, 2)
plt.plot(train_acc_list, label='train')
plt.plot(val_acc_list, label='val')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend()
plt.title('Accuracy')
```
